```python



"""
①platypusで個体生成する
②belegundu関数に入る
    ③varsには1個体分の設計変数が入る
    ③pop_vars_eval.txtにvars変数を書き込む　,をタブに変換する
    ③check_call(adress)でmazda_mop.exeを実行
    ③pop_objs_eval.txtを読み込み、objsリストに目的関数を入れる
    ③pop_cons_eval.txtを読み込み、consリストに制約点を入れる

"""
import time
import random
from platypus2 import MOEAD, Problem, Real, nondominated
import os
import sys
from subprocess import check_output
import numpy
import csv
import subprocess

if(os.path.isdir("data/mazda_pop") == False):
    os.mkdir("data/mazda_pop")


def text_csv_converter(datas): # datasはテキストファイルの場所
   # 保存するCSVファイルの場所
    file_csv = datas.replace("txt", "csv")
    # テキストファイルを開く
    with open(datas)as rf:
       # 書き込むＣＳＶファイルを開く
        with open(file_csv, "w")as wf:
            readfile = rf.readlines()
            for read_text in readfile:
               # listに分割
                read_text = read_text.split()
               # csvに書き込む
                writer = csv.writer(wf, delimiter=',')
                writer.writerow(read_text)



def belegundu(vars):  #目的関数の値のリストと制約条件関数の値のリストを返す

    #varsの222変数をpop_vars_eval.txtに書き込んで上書き
    
    if(os.path.isfile("Mazda_CdMOBP/Mazda_CdMOBP/src/vs2017/pop_vars_eval.txt")==True):
        os.remove("Mazda_CdMOBP/Mazda_CdMOBP/src/vs2017/pop_vars_eval.txt")
    with open("Mazda_CdMOBP/Mazda_CdMOBP/src/vs2017/pop_vars_eval.txt",'x') as f:           
            
        vars = str(vars)
        vars = vars.replace("'",'')
        vars = vars.replace("[",'')
        vars = vars.replace("]",'')
        vars = vars.replace(',','\t')      
        f.write(vars)
        
    with open("Mazda_CdMOBP/Mazda_CdMOBP/src/vs2017/pop_vars_eval.txt",'r') as f:
        s = f.read()

    cmd = "run.bat"
    subprocess.call(cmd, shell = True)
    
    text_csv_converter('D:/anaconda/envs/nakanishi/Mazda_CdMOBP/Mazda_CdMOBP/src/vs2017/pop_objs_eval.txt')
    with open('D:/anaconda/envs/nakanishi/Mazda_CdMOBP/Mazda_CdMOBP/src/vs2017/pop_objs_eval.csv') as f:
        reader = csv.reader(f,delimiter=',')
        data1 = [e for e in reader]

    
    objs = [data1[0][0],data1[0][1]]
    objs_f = [float(s) for s in objs]
    #objs_f[1] = -objs_f[1]
    
    
    
    
        
    #制約条件の関数値のref_pop_cons_eval.txtの一行とりだして54個中マイナスの数数える
    #マイナスの数だけ目的関数に1プラス

    text_csv_converter('Mazda_CdMOBP/Mazda_CdMOBP/src/vs2017/pop_cons_eval.txt')

    penalty = 0
    with open('Mazda_CdMOBP/Mazda_CdMOBP/src/vs2017/pop_cons_eval.csv') as f:
        reader = csv.reader(f,delimiter=',')
        data2 = [e for e in reader]
        
        #print("csvdata cons",data2)
        
    cons = []
    penalty = 0
    for i in range(54):
        cons.append(float(data2[0][i]))
        if(float(data2[0][i])<0):
            penalty +=1
    #print(objs_f, penalty)
            
    cons_f = [float(s) for s in cons]
    objs_f[0] = objs_f[0] + (penalty) 
    objs_f[1] = 74 + objs_f[1]
    
    
    
    return [objs_f[0], objs_f[1]], [penalty]

pop_size = 50  #個体数定義
gen_size = 100#世代数定義 
problem = Problem(222,2,1) #問題定義（設計変数,目的関数の数,制約の数）
problem.types[:] = [Real(0, 1)]*222 #設計変数の範囲
problem.directions[:] = [Problem.MINIMIZE,Problem.MINIMIZE] #各目的関数の目的
problem.constraints[:] = ">=0"
problem.function = belegundu #解く問題を設定


start_time = time.time()
algorithm = MOEAD(problem,population_size = pop_size)
algorithm.run(pop_size*gen_size*2)
execution_time = time.time() - start_time
print("処理時間")
print(execution_time)
```
